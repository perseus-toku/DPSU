```python
# generate a complex function according to some equations 
# then use a simple nn to predict the output of the function 
import numpy as np 
import torch 
# random noise must be used to the generation process 
import torch.nn.functional as F
# some ideas: just a multi linear model 
import os 
# apply a random noise to the input data 
EPSILON = 1e-5 

def generate_a_random_input(arr_len):
    arr = np.random.rand(1,arr_len)
    # multiply the number to increase the difference 
    rand_multiplier = np.random.randint(5)
    arr = arr * rand_multiplier
    return arr 

def generate_k_samples(k, arr_len):
    samples =[]
    for i in range(k):
        samples.append(generate_a_random_input(arr_len))
    return np.array(samples)

num_train_samples = 1500 
num_test_samples = 500 
num_variables = 10 


def generate_labels(samples):
    # apply a given function to the input samples
    labels = [] 
    for s in samples:
        # step 1: exp 
        s = np.exp(s)
        # step 2: square 
        s = s*s
        #step 3: sum 
        s = np.sum(s)
        label = s
        labels.append(label)
    return labels 

train_samples = generate_k_samples(num_train_samples, num_variables)
test_samples = generate_k_samples(num_test_samples, num_variables)
# labels
train_labels = generate_labels(train_samples)
test_labels = generate_labels(test_samples)
```

```python
def data_transformation1(samples):
    new_samples = [] 
    for sample in samples:
        sample = np.exp(sample) 
        new_samples.append(sample)
        #sample = np.exp(sample)
    return np.array(new_samples)

```

```python
# apply a more complete transformation 
def data_transformation2(samples):
    new_samples = [] 
    for sample in samples:
        sample = np.exp(sample) 
        sample = sample * sample
        new_samples.append(sample)
        #sample = np.exp(sample)
    return np.array(new_samples)
```

```python
class Net(torch.nn.Module):
    def __init__(self, num_inputs, num_outputs, num_transformations): 
        # num Transformations defines the number of data transformations we have 
        # classification --> task specific 
        super(Net, self).__init__()
        
        self.fc1 = torch.nn.Linear(num_inputs,64)
        self.fc2 = torch.nn.Linear(64,num_outputs)
        #self.fc3 = torch.nn.Linear(32,num_outputs)
        # there shud be a torch.sigmoid for the gate value --> squeeze to [0,1] 
        
    def forward(self,x):
        # a very simple network structure 
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        #x = self.fc3(x)
        return x 
        
        
        
        
def train_network(train_samples, train_labels, test_samples, test_labels,NUM_EPOCH=50, save_model_idx = None, read_model_idx= None):
    print("start training...")
        
    net = Net(num_variables, 1)
    criterion = torch.nn.MSELoss()
    optimizer = torch.optim.Adam(net.parameters(), lr= 0.0001 )

    if read_model_idx:
        print("Loading an existing model with idx=", read_model_idx)
        assert os.path.exists("models"), "no saved model found"
        path = os.path.join("models", str(read_model_idx))
        net.load_state_dict(torch.load(path))

    device = torch.device("cuda:0")
    net = net.to(device)
    
    train_process = []
    test_process = []

    ## Train the given neural network 
    for e in range(NUM_EPOCH):
        print("Epoch ",e)
        print("-"*10)
        running_loss = 0.0 
        for i in range(len(train_samples)):
            optimizer.zero_grad()
            sample, label = train_samples[i], train_labels[i]
            sample = torch.tensor(sample) 
            sample = sample.type('torch.FloatTensor')
            label = torch.tensor(label)
    
            sample = sample.to(device)
            label = label.to(device)

            output = net(sample)
            loss = criterion(output, label)
            # get the gradients 
            loss.backward()
            running_loss += loss 
        
            optimizer.step()
    
        # evaluate on test set 
        running_test_loss =0.0 
        for i in range(len(test_samples)):
            sample, label = test_samples[i], test_labels[i]
            sample = torch.tensor(sample) 
            sample = sample.type('torch.FloatTensor')
            label = torch.tensor(label)
    
            sample = sample.to(device)
            label = label.to(device)

            output = net(sample)
            loss = criterion(output, label)
            running_test_loss += loss
    
        running_loss = running_loss/len(train_samples)
        running_test_loss = running_test_loss/len(test_samples)
    
        train_process.append(running_loss)
        test_process.append(running_test_loss)
        print("train loss is ",running_loss.tolist(), "test loss is ",running_test_loss.tolist())
    print("finished training")
                  
                  
    # save the model 
    if not os.path.exists("models"):
        os.mkdir("models")
    path = os.path.join("models", str(save_model_idx))
    torch.save(net.state_dict(),path)
    return train_process,test_process
    

        
```

```python
# make the network decide which one is the best data transformation by using the gates 
# the input format should be [df1, df2, .., dfn] for n potential data transformations 
# then there are n different gates --> G1, .., Gn for each point 
import torch 
from torch.autograd import Variable

class PreprocessingValidator(torch.nn.Module):
    def __init__(self, num_inputs, num_outputs, num_transformations): 
        # num Transformations defines the number of data transformations we have 
        # classification --> task specific 
        super(PreprocessingValidator, self).__init__()
        
        # first layer is a list of gates 
        self.gates = [] 
#         for i in range(num_transformations):
#             #G = torch.tensor(100.0, requires_grad = True)
#             self.G = Variable(torch.randn(1)*5, requires_grad=True)
#             self.gates.append(G)
#         self.gates = torch.tensor(self.gates, requires_grad=True).cuda()
#         self.single_gate = torch.nn.Parameter(torch.tensor(100.0), requires_grad = True).cuda()

        # large scale data transformation --> data are divided to batches and somehow 
        # the dimension of the data can be reduced by the network --> to give a ranking of given data 
        # preprocessing being done 
        # layer 1 size increases linearly with the number of transformations 
        # use gates to multiply first 
        self._first_layer_width = num_inputs * num_transformations
        self.single_input_width = num_inputs
        
        self.fc1 = torch.nn.Linear(num_inputs * num_transformations ,64)
        self.fc2 = torch.nn.Linear(64 ,32)
        self.fc3 = torch.nn.Linear(32,num_outputs)
#         self.scale = torch.nn.Parameter(torch.ones(1).cuda())

        
        self.G1 = torch.nn.Parameter((torch.ones(1)).cuda(), requires_grad = True)
        self.G2 = torch.nn.Parameter((torch.ones(1)).cuda(), requires_grad = True)
        self.G3 = torch.nn.Parameter((torch.ones(1)).cuda(), requires_grad = True)
        self.a1 = None
        self.a2 = None 
        self.a3 = None 
        
    def forward(self,x):
        # a very simple network structure 
        # is a tensor --> multi with the gates --> index with the tesor 
        # there shud be a torch.sigmoid for the gate value --> squeeze to [0,1] 
#         nxt = [] 
#         for i in range(0, len(self.gates)):
#             new_x = x[i,:] * torch.sigmoid(self.gates[i]) 
#             nxt.append(new_x)
#         x = torch.cat((nxt), 0)

        # essentially a softmax layer 
        sig_g1 = torch.sigmoid(self.G1)
        sig_g2 = torch.sigmoid(self.G2)
        sig_g3 = torch.sigmoid(self.G3)

        norm_factor = torch.exp(sig_g1) + torch.exp(sig_g2) + \
        torch.exp(sig_g3)
        
        self.a1 = torch.exp(sig_g1)/norm_factor
        self.a2 = torch.exp(sig_g2)/norm_factor
        self.a3 = torch.exp(sig_g3)/norm_factor

            
        x  = torch.cat((x[0,:]* self.a1, x[1,:]* self.a2, x[2,:]* self.a3), 0 )

#         x  = torch.cat((x[0,:]* torch.sigmoid(self.G1),  x[1,:]* torch.sigmoid(self.G2),  x[2,:]* torch.sigmoid(self.G3)), 0 )
        # concatenate the new x together --> gates are optimized this way 
        x = x.reshape(-1)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        #x = self.fc3(x)
        return x 


```
